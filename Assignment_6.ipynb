4. Plotting a basic sine graph by adding more features. Adding Multiple plots by Superimposition like cosine wave.

```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
z = np.cos(x)
plt.xlabel("Angle")
plt.ylabel("Magnitude")
plt.title("Sine/Cosine function curve")
plt.plot(x,y,'b--', x,z,'r:')
plt.show()
```

6.  Plot Histogram, We have a sample data of Students marks of various Students, we will try to plot number of Students by marks range and try to figure out how many Students are average, below-average and Excellent.

Marks = [ 61,86,42,46,73,95,65,78,53,92,55,69,70,49,72,86,64]

Histogram showing Below Average, Average and Execellent distribution

40-60: Below Average

60-80: Average

80-100: Excellent

```python
Marks = np.array([ 61,86,42,46,73,95,65,78,53,92,55,69,70,49,72,86,64])
excellent = Marks[np.logical_and(Marks >= 80, Marks < 100)]
average = Marks[np.logical_and(Marks >= 60, Marks < 80)]
below_avg = Marks[np.logical_and(Marks >= 40, Marks < 60)]
plt.xlabel('Marks')
plt.ylabel('Students')
plt.title('Histogram of Marks')
plt.hist(Marks, color='orange', edgecolor='black')
plt.show()
```

7.  Titanic Data Set Download Data

 Load the data file

 (i) Create a pie chart presenting the male/female proportion 

 (ii) Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
import pandas as pd
df = pd.read_csv("/content/titanic.csv")
df
```

```python
df.columns
```

```python
cnt = df.Sex.value_counts()
plt.pie(cnt, labels = ["Male", "Female"], shadow = True, autopct = "%0.0f%%")
plt.legend()
plt.title("Male / Female Proportion")
plt.show()
```

```python
df_male_count = df[df["Sex"] == "male"]
df_male_count.Sex.value_counts()
```

```python
df_female_count = df[df["Sex"] == "female"]
df_female_count.Sex.value_counts()
```

```python
plt.scatter(df_male_count.Fare, df_male_count.Age, label = "Male", marker = "^", color = "orange")
plt.scatter(df_female_count.Fare, df_female_count.Age, label = "Female", marker = "*", color = "b")
plt.title("Fare Paid and Age by Gender")
plt.xlabel("Fare")
plt.ylabel("Age")
plt.legend()
plt.show()
```
