# Exploratory Data Analysis

```python
df['Survived'].value_counts()
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Survived', data=df)
```

```python
df1 = df['Survived'].value_counts()
plt.pie(df1.values, labels=df1.index, autopct='%0.2f%%', colors='rg')
plt.title('Pecentage of Survived Yes/No', fontsize=15)
plt.show()
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(), yticklabels='False')
```

```python
import missingno as no
no.bar(df, color='y')
plt.show()
```

## Univariante Analysis

```python
sns.distplot(df['Age'], color='r')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='YlGn')
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=df)
plt.show()
```

```python
sns.boxplot(x='Survived', y='Fare', data=df)
plt.show()
```

```python
sns.boxplot(x='Pclass',y='Age',data=df)
plt.show()
```

```python
sns.FacetGrid(df, hue='Survived').map(sns.distplot, 'Age').add_legend()
```

## Bivariant Analysis

```python
sns.set_style("whitegrid")
sns.FacetGrid(df, hue="Survived").map(plt.scatter, 'Age','Parch').add_legend()
```
